```python
def convert(n):
    if n==1:
        return "yes"
    return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(convert)
```

```python

ax = sns.countplot(x = 'Churn' , data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
gb
```

```python
plt.pie(gb['Churn'],labels= gb.index,autopct = "%1.2f%%")
plt.title("percantage of Customers churned")
plt.show()
```

26.54% of Customers are no longer associated with company , lets find out why?

```python
plt.figure(figsize = (5,4))
sns.countplot(x = 'gender', data=df,hue = "Churn")
plt.title("Churned Customer's Gender")
plt.show()
```

```python
ax = sns.countplot(x = 'SeniorCitizen' , data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers as Seniors")
plt.show()
```

```python
cross = pd.crosstab(df['SeniorCitizen'], df['Churn'])


cross_pct = cross.div(cross.sum(axis=1), axis=0) * 100


colors = ['#1f77b4', '#ff7f0e']  

cross_pct.plot(kind='bar', stacked=True, figsize=(6, 4), color=colors)


for i, row in enumerate(cross_pct.values):
    bottom = 0
    for j, val in enumerate(row):
        if val > 0:
            plt.text(i, bottom + val / 2, f"{val:.1f}%", ha='center', va='center', color='white', fontsize=10)
        bottom += val


plt.title('Churn by Senior Citizen Status (in %)')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

comparatively, greater percent of senior citizens have churned.

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x="Contract",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Number of Customers in Each Contract")
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies']
c = 3
r = (len(columns)//c)

plt.figure(figsize = (c*5,r*4))
for idx,col in enumerate(columns,1):
    plt.subplot(r,c,idx)
    sns.countplot(x=col,data=df,hue="Churn")
    plt.title(f'{col} by Churn')
    
   

plt.tight_layout()
plt.show()
    
```

The majority of Customers who tend not to churn had services like 'PhoneService','InternetService'(DSL) and Online Security enabled.
For Services like OnlineBackup , TechSupport and StreamingTV , churn rates are noticably higher when these services were not used or were unavailable.
